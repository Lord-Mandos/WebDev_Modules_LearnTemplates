>## **<span style="color:#6be7fa">CSS</span> Selectors**

> ### **Element Selector**

zielt auf alle HTML Tags unter dem Namen im gesamten Dokument



>### **Class Selector**

wird mit einem Punkt initialisiert, hier kann jeder beliebige Name vergeben werden\
sprechende Variablennamen

ein Class Selector wird eingesetzt um nur bestimmten Elementen eine Gruppe von Eigenschaften mit zu geben,\
indem der Class Selector einem html Tag als Attribut bergeben wird



> ### **Id Selector**

hnliche Funktionsweise wie der Class Selector,\
jedoch kann der Id Selector nur auf ein einziges HTML Tag angewandt\
und dann nicht mehr eingesetzt werden.\
Dadurch wird eine hierarchische Struktur erzeugt, Id Selectoren haben immer Vorrang.



>### **Attribute Selector**
bezieht sich auf die HTML Elemente , die mit einem bestimmten Attribut versehen wurden



> ### **Universal Selector**
Bezieht sich auf alle HTML Tags
