> ## **<span style="color:#6be7fa">CSS</span> Positioning**



[Positioning](https://appbrewery.github.io/css-positioning/)

> ### **Static Positioning**

* HTML default flow
* alles was wir bisher gesehen haben

> ### **Relative Positioning**

* positioniert ein Objekt relativ zur Original- bzw. Standardposition (Static Position)
* es knnen also nderungen relativ zur static position vorgenommen werden
* das Objekt verschiebt dabei keine anderen Elemente auf dem Bildschirm

> ### **Absolute Positioning**

* Position relativ zum nchstgelegenen bergeordneten Element oder der linken oberen Ecke der Webseite
* bergeordnetes Element hat in der Regel die Position Property auf "relative" gesetzt
* hat kein bergeordnetes Element die Position Property gesetzt, so wird das Element relativ zur Webseite positioniert

> #### **Z-<span style="color:#6be7fa">Index</span>**

* definiert in welcher Ebene sich das Element auf der Seite befindet
* je hher der Wert desto weiter oben befindet sich das Element

> ### **Fixed Positioning**

* relative Position zur oberen linken Ecke des Browserfensters
